```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
file = 'data/frmgham2.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="UTF-8")
```

```python
# Preview of the DataFrame
# Note that FIELD8 is likely a meaningless column
df.head()
```

```python
# Clean to columns needed
clean_df = df[['RANDID','SEX','AGE','SYSBP','DIABP','CURSMOKE','BMI','DIABETES','BPMEDS','HEARTRTE','GLUCOSE','educ','STROKE','CVD','HYPERTEN']]
clean_df.head()
```

```python
# Identify incomplete rows
clean_df.count()
```

```python
# Drop all rows with missing information
clean2_df = clean_df.dropna(how='any')
```

```python
# Verify dropped rows
clean2_df.count()
```

```python
clean2_df.to_csv("data/max_reduce.csv")
```
